```python
#Q1
import numpy as np

arr = np.random.randint(10, 101, (5, 6))

print("2D Array:")
print(arr)

# Print Shape
print("\nShape:", arr.shape)

# Print Total Number of Elements
print("Size:", arr.size)

# Print Data Type
print("Data Type:", arr.dtype)
```

```python
#Q2
import numpy as np
arr = np.random.randint(1, 51, 20)

print("Array:")
print(arr)

# Minimum value and index
print("\nMinimum Value:", np.min(arr))
print("Index of Minimum Value:", np.argmin(arr))

# Maximum value and index
print("\nMaximum Value:", np.max(arr))
print("Index of Maximum Value:", np.argmax(arr))

# Sum
print("\nSum:", np.sum(arr))

# Mean
print("Mean:", np.mean(arr))

# Standard Deviation
print("Standard Deviation:", np.std(arr))
```

```python
#Q3
import numpy as np
# Create a 4x5 matrix of random integers
matrix = np.random.randint(20, 81, (4, 5))

print("Matrix:")
print(matrix)

# Entire Matrix Statistics
print("\nMinimum Value:", np.min(matrix))
print("Maximum Value:", np.max(matrix))

print("Sum:", np.sum(matrix))

print("Mean:", np.mean(matrix))

print("Standard Deviation:", np.std(matrix))

# Sum of each row
print("\nRow-wise Sum:")
print(np.sum(matrix, axis=1))

# Sum of each column
print("\nColumn-wise Sum:")
print(np.sum(matrix, axis=0))
```

```python
#Q4
import numpy as np
# Create array from 1 to 24
arr = np.arange(1, 25)

print("Original Array:")
print(arr)

# Reshape into (4,6)
array1 = arr.reshape(4, 6)

print("\nReshape (4,6):")
print(array1)
print("Shape:", array1.shape)

# Reshape into (3,8)
array2 = arr.reshape(3, 8)

print("\nReshape (3,8):")
print(array2)
print("Shape:", array2.shape)

# Reshape into (2,3,4)
array3 = arr.reshape(2, 3, 4)

print("\nReshape (2,3,4):")
print(array3)
print("Shape:", array3.shape)
```
